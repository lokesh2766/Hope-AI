**Q1. Create a class and function, and list out the items in the list (AI Subfields)**

**Q2. Create a function that checks whether the given number is Odd or Even**

```python
class OddEven:
    def OddEven(self):
        num = int(input("Enter a number: "))
        if num % 2 == 0:
            print(num, "is Even number")
        else:
            print(num, "is Odd number")

obj2 = OddEven()
obj2.OddEven()
```

**Q3. Tells eligibility of marriage for male and female (21 for male, 18 for female)**

```python
class ElegiblityForMarriage:
    def Elegible(self):
        gender = input("Your Gender:")
        age = int(input("Your Age:"))
        if gender == "Male" and age >= 21:
            print("ELIGIBLE")
        elif gender == "Female" and age >= 18:
            print("ELIGIBLE")
        else:
            print("NOT ELIGIBLE")

obj3 = ElegiblityForMarriage()
obj3.Elegible()
```

**Q4. Calculate the percentage of your 10th mark**

```python
class FindPercent:
    def percentage(self):
        s1 = int(input("Subject1= "))
        s2 = int(input("Subject2= "))
        s3 = int(input("Subject3= "))
        s4 = int(input("Subject4= "))
        s5 = int(input("Subject5= "))
        total = s1 + s2 + s3 + s4 + s5
        percent = (total / 500) * 100
        print("Total :", total)
        print("Percentage :", percent)

obj4 = FindPercent()
obj4.percentage()
```

**Q5. Print area and perimeter of a triangle using class and functions**

```python
class Triangle:
    def area(self):
        height = int(input("Height:"))
        breadth = int(input("Breadth:"))
        print("Area formula: (Height*Breadth)/2")
        area = (height * breadth) / 2
        print("Area of Triangle:", area)

    def perimeter(self):
        h1 = int(input("Height1:"))
        h2 = int(input("Height2:"))
        b = int(input("Breadth:"))
        print("Perimeter formula: Height1+Height2+Breadth")
        perimeter = h1 + h2 + b
        print("Perimeter of Triangle:", perimeter)

obj5 = Triangle()
# call whichever you need:
obj5.area()
# obj5.perimeter()
```
